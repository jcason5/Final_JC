```python
import random #used to randomize the word chosen for hangman

#funtion used to get random word from the list
def get_word():
    words_list = ['Variable', 'List', 'Tuple', 'Dictionary', 'Function', 'Class', 'Inheritance',
                  'Module', 'Package', 'Loop', 'Conditional', 'Exception', 'Lambda', 'Decorator',
                  'Generator', 'Slice', 'Recursion', 'API', 'Framework', 'IDE']
    word = random.choice(words_list)
    return word.lower()

# Function to display Hangman images 
def display(attempts):
    stages = [""" 
                --------
                |      |
                |      o
                |     \|/
                |      |
                |     / \\
                _
                """,
                """  
                --------
                |      |
                |      o
                |     \|/
                |      |
                |     /
                _
                """,
                """  
                --------
                |      |
                |      o
                |     \|/
                |      |
                |     
                _
                """,
                """  
                --------
                |      |
                |      o
                |      |/
                |      |
                |     
                _
                """,
                """  
                --------
                |      |
                |      o
                |      |
                |      |
                |     
                _
                """,
                """  
                --------
                |      |
                |      o
                |      |
                |      
                |     
                _
                """,
                """  
                --------
                |      |
                |      
                |      
                |      
                |     
                _
                """
    ]
    return stages[attempts]

# This is the function with the actual hangman game
def play():
    # welcoming messages
    print("Welcome to Hangman!")
    print("The category is Python")
    print("You have 6 attempts to guess the word.")
    # empty lists needed to collect used letters/words
    guessed_letters = []
    guessed_words = []
    word_to_guess = get_word()
    attempts = 6
    guessed_word = "_ " * len(word_to_guess)

    # Game code continues the game till all letters are guessed
    while True:
        print(guessed_word)
        print(display(attempts))

        guess = input("Guess a letter or word: ").lower()

        if guess in guessed_letters:
            print("You already guessed that letter. Try again.")
            continue

        if len(guess) == 1 and guess.isalpha():
            if guess in word_to_guess:
                for index, letter in enumerate(word_to_guess):
                    if letter == guess:
                        guessed_word[index] = ''.join(guess)
            else:
                attempts -= 1
                print("Incorrect guess! Attempts remaining:", attempts)
        elif len(guess) == len(word_to_guess) and guess.isalpha():
            if guess in guessed_words:
                print("You've already guessed that word.")
            elif guess != word_to_guess:
                print(f"{guess} is not the word")
                attempts -= 1
                guessed_words.append(guess)
            else:
                print("Congratulations! You guessed the word!")
                break

        if attempts == 0:
            print("Sorry, you ran out of attempts. The word was:", word_to_guess)
            print(display(attempts))
            break
        elif "_" not in guessed_word:
            print("Congratulations! You guessed the word!")
            break

if __name__ == "__main__":
    play()
```
